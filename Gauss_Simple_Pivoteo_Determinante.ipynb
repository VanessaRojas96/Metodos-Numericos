```python
vector,matrix,x,producto=[],[],[],1.0
m=int(raw_input('Valor de m:'))
n=int(raw_input('Valor de n:'))
for z in range(0,n):
    x.append(n*0.0)

if(m==n):
    for i in range(m):#filas
        matrix.append([])
        for j in range(n):#columnas
            matrix[i].append(0)

    for k in range(n):
        vector.append(0)
    
    print 'Introduce la matriz de coeficientes y el vector soluci칩n'
    for z in range(0,m):
        for y in range(0,n):
            matrix[z][y]=float((raw_input("\nElemento a"+str(z+1)+""+str(y+1)+": ")))
        vector[z]=float((raw_input('\nSolucion '+str(z+1)+': ')))
    print "\nLa matriz original es: ",matrix
    print "\nEl vector solucion original es: ",vector
    
    p,k=0,0
    p=k
    big = abs(matrix[k][k])
    for i in range(k+1,n):
        dummy = abs(matrix[i][k])
        if (dummy>big):
            big=dummy
            p=i
    if (p!=k):
        for j in range(k,n):
            dummy=matrix[p][j]
            matrix[p][j]=matrix[k][j]
            matrix[k][j]=dummy
        dummy=vector[p]
        vector[p]=vector[k]
        vector[k]=dummy
    
    for z in range (0,m):
        for y in range(z+1,m):
            factor=(matrix[y][z]/matrix[z][z])
            vector[y]=vector[y]-(factor*vector[y])
            for w in range(0,n):
                matrix[y][w]=matrix[y][w]-(factor*matrix[z][w])
    #sustituci칩n hacia atr치s
    x[m-1]=vector[m-1]/matrix[m-1][m-1]
    
    for y in range(m-2,-1,-1):
        suma=0
        for w in range(0,n):
            suma=suma+matrix[y][w]*x[w]
        x[y]=(vector[y]-suma)/matrix[y][y]
        
    print '\nResultado de la matriz\n', matrix
    print '\nResultado del vector\n', vector
    print '\nRespuestas de las incognitas:\n', x
    for i in range(len(x)):
        print "\nX%d=%f" %(i+1,x[i])
    
    for i in range (0,n):
        for j in range (0,n):
            if(i==j):
                producto=producto*matrix[i][j]
    print "El determinante es= ",producto
else:
    print "No se puede usar Gauss, no es una matriz cuadrada"
```
